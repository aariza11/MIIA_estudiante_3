```python
# código
import numpy as np
#FUNCION GENERAL
def total(lista,num):
    h=[]
    j=1
    c=0
    #temp=0
    #se verifica si el tamaño de la lista de promedios es mayor al numero de becas
    if len(lista)>num:
        
        while j<=num:
            temp=0
            u=np.quantile(lista,0.98)
            for i in range(len(lista)):
                if lista[i]>u:
                    c=lista[i]
                    temp=i
            
            h.append(c)
            lista.pop(temp)
            #h.append(c)
            j+=1
    else:
        
        while j<=len(lista):
            temp1=0
            u1=np.quantile(lista,0.98)
            for k in range(len(lista)):
                if lista[k]>u1:
                    c=lista[k]
                    temp1=k
            
            h.append(c)
            lista.pop(temp1)
    
    return h

#funcion que retorna un diccionario con los promedios becados por estrato
def opcion1(estrato,notas,becas):
    notas_estrato5=[]
    notas_estrato4=[]
    notas_estrato3=[]
    notas_estrato2=[]
    notas_estrato1=[]
    
   
    #se extraen los datos de la base grande
    for i in range(len(estrato)):
        if estrato[i]==5:
            notas_estrato5.append(notas[i])
            
    
    for i in range(len(estrato)):
        if estrato[i]==4:
            notas_estrato4.append(notas[i])
    
    for i in range(len(estrato)):
        if estrato[i]==3:
            notas_estrato3.append(notas[i])
    
    for i in range(len(estrato)):
        if estrato[i]==2:
            notas_estrato2.append(notas[i])
    
    for i in range(len(estrato)):
        if estrato[i]==1:
            notas_estrato1.append(notas[i])
    
    #se asignan las becas a los mejores puntajes
    becas_estrato_5=total(notas_estrato5,becas)
    becas_estrato_4=total(notas_estrato4,becas)
    becas_estrato_3=total(notas_estrato3,becas)
    becas_estrato_2=total(notas_estrato2,becas)
    becas_estrato_1=total(notas_estrato1,becas)
    
    #valores que se retornaran en forma de diccionario
    estratos=['Estrato 5','Estrato 4','Estrato 3','Estrato 2','Estrato 1']
    notas_conglomerado=[]
    notas_conglomerado.append(becas_estrato_5)
    notas_conglomerado.append(becas_estrato_4)
    notas_conglomerado.append(becas_estrato_3)
    notas_conglomerado.append(becas_estrato_2)
    notas_conglomerado.append(becas_estrato_1)
    
    #valor a retornar
    dic = dict(zip(estratos,notas_conglomerado))
    
    
    
    return dic
    
print(opcion1(estrato,promedio,6))


```
